_2- R:_ *Sim, os resultados fazem sentido. As colunas de valor da fatura (BILL_AMT1 a BILL_AMT6) apresentam:*

Exercício 3: Visualize as características de valor da fatura usando uma grade 2 por 3 de plotagens de histograma. Dica: Você pode usar 20 bins para essa visualização.

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i, feature in enumerate(fatura):
    axes[i].hist(df[feature], bins=20, color='blue', alpha=0.7)
    axes[i].set_title(f'Histograma de {feature}')
    axes[i].set_xlabel('Valor da Fatura')
    axes[i].set_ylabel('Frequência')

plt.tight_layout()
plt.show()
```

Exercício 5: Plote um histograma das características de pagamento da fatura semelhante ao das características de valor da fatura, mas aplique também alguma rotação aos rótulos do eixo x com o argumento de palavra-chave xrot para que eles não se sobreponham. Podemos incluir o argumento de palavra-chave xrot=<ângulo> em qualquer função de plotagem para girar os rótulos do eixo x de acordo com um ângulo específico em graus. Considere os resultados.

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i, pay in enumerate(pagamento):
    
    df_filtrado = df[df[pay] < 30000]
    
    axes[i].hist(df_filtrado[pay], bins=20, color='green', alpha=0.7)
    axes[i].set_title(f'Histograma de {pay}')
    axes[i].set_xlabel('Valor do Pagamento')
    axes[i].set_ylabel('Frequência')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

O uso da máscara booleana confirmou isso ao revelar que milhares de registros têm valor de pagamento igual a zero em cada mês.*

*Esse comportamento pode estar associado a:*

*Clientes que usaram o cartão pouco e por isso não tinham valor a pagar.*

*Possíveis atrasos ou estratégias de pagamento alternadas.*

*É coerente com a realidade do crédito rotativo, onde parte dos clientes não realiza pagamento total todos os meses.*

Exercício 7: Ignorando os pagamentos iguais a 0 usando a máscara que criou na etapa anterior, utilize o método .apply() do pandas e o método np.log10() do NumPy para plotar histogramas de transformações logarítmicas dos pagamentos diferentes de zero. Considere os resultados. Dica: Você pode usar .apply() para aplicar qualquer função, inclusive log10, a todos os elementos de um DataFrame ou de uma coluna usando a sintaxe a seguir: .apply(<nome_função>).

```python
import numpy as np
df_pagamentos_positivos = df[pagamento][df[pagamento] > 0]

df_log_pagamentos = df_pagamentos_positivos.apply(np.log10)

plt.figure(figsize=(15, 10))
for i, pay in enumerate(pagamento):
    plt.subplot(2, 3, i + 1)
    df_log_pagamentos[pay].dropna().plot.hist(bins=20, edgecolor='black')
    plt.title(f'Log10 - {pay}')
    plt.xlabel('log10(valor)')
    plt.ylabel('Frequência')
plt.tight_layout()
plt.show()
```
